# Create dataframes using list

```python
name=['Suresh','Satish','Siva']
age=[32,30,28]
name,age
```

```python
pd.DataFrame(zip(name,age))
```

```python
data=zip(name,age)
col=['Name','Age']
pd.DataFrame(data,columns=col)
```

$ Providing Index $

```python
data=zip(name,age)
col=['Name','Age']
ind=['A','B','C']
pd.DataFrame(data,columns=col,index=ind)
```

$ Add New Column $

```python
data=zip(name,age)
col=['Name','Age']
ind=['A','B','C']
df=pd.DataFrame(data,columns=col,index=ind)
df
```

```python
city_names=['Blr','Hyd','Amr']
df['city']=city_names
df
```

$update the existing columns$

```python
df['Name']=['Hari','Giri','Cherry']
df
```

Drop the Column 

```python
df.drop('city',axis=1,inplace=True)
```

```python
df
```

Drop the Row

```python
df.drop('C',axis=0,inplace=True)
```

```python
df
```

$ Save the dataframe$

```python
df.to_csv("output.csv")
df.to_excel("output.xlsx")
```
